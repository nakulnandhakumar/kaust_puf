```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
# Assuming your dataset is in a CSV format with columns "Time" and "Intensity"
df = pd.read_csv('puf_dataset_07_08/Can-D1-50mA.csv')

# Extract "Time" and "Intensity" columns
time = df['Time'].values[:1000]  # Take the first 10,000 points
intensity = df['Intensity'].values[:1000]

# Adjust the time axis: start from 0 and convert to picoseconds (ps)
# Original time axis is from -1e-4 to 1e-4 sec, with 10 ps apart
time_adjusted = (time - np.min(time)) * 1e12  # Convert to picoseconds (ps)

# Plot the first 10,000 points
plt.figure(figsize=(10, 6))
plt.plot(time_adjusted, intensity)

# Set up labels and title
plt.xlabel('Time (ps)', fontsize=20, fontfamily='Arial')
plt.ylabel('Intensity', fontsize=20, fontfamily='Arial')

# Set tick labels with only 10 labels on the x-axis
plt.xticks(np.arange(0, np.max(time_adjusted)+1, 1000), fontsize=16)  # Step by 10,000 ps
plt.yticks(fontsize=16)

# Save figure
plt.savefig('figures/laser_signal/chaotic_laser_subset_10k_points.png', dpi=300)
```
